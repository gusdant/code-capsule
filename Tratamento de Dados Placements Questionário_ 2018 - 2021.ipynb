```python
#Importando Pacotes 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import openpyxl
from pathlib import Path
```

### Abrindo e visualizando tabela 

& Aplicando para placementID

```python
#Abrindo a tabela 2021 Excel
dataframe = pd.read_excel("DataFrameVagas_Full.xlsx", header = 0, index_col= 0)

#Visualizando
display(dataframe)
```

```python
dataframe.dtypes
#Visualizando os tipos de dados presentes na tabela
```

```python
#Agora, removeremos rows que tenham email_primario, email_secundario, telefone_celular, telefone_principal
#Para isso vamos substituir os valores NaN de tais colunas por valores do formato "string":
#Substituindo

dataframe['email_primario'] = dataframe['email_primario'].fillna('NA')
dataframe['email_secundario'] = dataframe['email_secundario'].fillna('NA')
dataframe['telefone_celular'] = dataframe['telefone_celular'].fillna('NA')
dataframe['telefone_principal'] = dataframe['telefone_principal'].fillna('NA')

dataframe
```

```python
#montando um dataframe com os placements sem contato

apagar_df = dataframe[(dataframe['email_primario'] == 'NA') & (dataframe['email_secundario'] == 'NA') & (dataframe['telefone_celular'] == 'NA') & (dataframe['telefone_principal'] == 'NA')]

apagar_df.head(90)
```

```python
#Agora iremos tentar apagar as rows com IDs de placement que estão no dataframe apagar_df
#Para isso vamos criar uma lista com placements id desse dataframe

IDapagar_list = apagar_df["placement_id"].tolist()
IDapagar_list

```

```python
#Agora apagando IDs repetidos da lista de placement_ids que iremos deletar do dataframe

lst = list()
lst = [0]
repetidos = list()
for ID in IDapagar_list:
    if ID in lst:
      print('repetido')
      repetidos.append(ID)
    else:
      lst.append(ID)      

lst
```

```python
#Transformando a IDapagar_list na lista sem número repetidos
IDapagar_list = lst

#Usando IDapagar_list para apagar as vagas que não possuem os dados desejados
#Checando o formato inicial do dataframe
dataframe.shape

#Aplicando o filto para conseguir o data frame final filtrado
filtro = dataframe[~dataframe['placement_id'].isin(IDapagar_list)]

#checando alguns dos dados
filtro.head(100)


```

```python
#Atribuindo um nome para o novo dataframe final filtrado
filtrado_df = filtro
filtrado_df
```

```python
#Passando para Excel

filtrado_df.to_excel('Placements_Questionario_Filtrado', sheet_name = 'pagina01', index = True, engine = 'xlsxwriter')
```

### Aplicando para Vaga IDs repetidos


```python
#Consertando a repetição de Id_Vagas
#Usaremos o mesmo tratamento que usamos para id_placement exceto que trataremos
#com index 

vaga_apagar_lst = apagar_df.index.tolist()
vaga_apagar_lst
```

```python
#Agora apagando Vaga_IDs repetidos
#Assim ficaremos com a lista de vagas_ID que precisamos apagar

vaga_list = list()
vaga_list = [0]
vaga_repetidos = list()
for ID in vaga_apagar_lst:
    if ID in vaga_list:
      print('repetido')
      vaga_repetidos.append(ID)
    else:
      vaga_list.append(ID)      
vaga_list
```

```python
#Agora iremos criar um dataframe com o filtro dos vaga_id que precisamos apagar
filtro_vagaid_df = dataframe[~dataframe.index.isin(vaga_apagar_lst)]

filtro_vagaid_df
```

```python
#Passando para Excel (não vamos passar agora)

filtro_vagaid_df.to_excel('Placements_NaoFinalizado_Filtrado_VagaID', sheet_name = 'pagina01', index = True, engine = 'xlsxwriter')
```
